#### Domanda 5
**Suppose you use Batch Gradient Descent and you plot the validation error at every epoch. Suppose the validation error consistently goes up. what is likely going on? How can you fix this?**

The model is overfitting the data. This means that the model is starting to minimize just the training error and at the same time increasing the test error (whose estimation is given by the validation error). In this case I would give the model moore data or on the other hand consider a less complex model. 
Making the model less complex will help reducing the model variance and increasing the model bias, thus, hopefully, achieving a better bias-variance trade-off.

**Is it a good idea to stop Mini-Batch gradient descent immediately when the validation error goes up?**

No it is not. Infact the mini-batch gd is a stochastic algorithm. One iteration is stochastic and its result may just depend upon case. Stopping immediately the algorithm after an increase in validation error, would not be a good idea. It could just be the resul of a bad round of luck. Maybe the algorithm could have achieved an even bettere result.

**Suppose you are using polynomial regression. You plot the learning curve and there is a large gap between the training error and the validation error. What is happening? What are three ways to solve it?**

It means that the model is overfitting. Having higher validation error than training error means that the model has lost generality. In this case you can:
1. Create a less complex model
2. Feed the model more data to help the training part and permit an optimal value for the parameters
3. Create different features, that are more informative and realted to the response variable.
